```python
for t in range(100,2000,100):
    print("elaborazione epoca n: ",t)
    Xi = np.load(path + 'model' + '_' + str(t) + 'epochs' + '.npy')
    Xi = Xi.reshape((-1,1))
    xsize = 8
    ysize = 5
    plt.figure(figsize=(xsize, ysize))
    plt.title('Distribuzione dei coefficienti, t: '+str(t))
    plt.hist(Xi,bins = 20)
    plt.xlabel('$xi$', labelpad = 10)
    #plt.xlim((0.1,2))
    #plt.ylim((0,20))
    
    plt.savefig(path + 'plot/Xi_distribution_'+str(t)+'.png', dpi = 300, pad_inches = .1, bbox_inches = 'tight')
```
