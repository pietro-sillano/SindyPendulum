```python
import pandas as pd
import csv
import matplotlib.pyplot as plt
```

```python
path = '../model/' # con regolarizzatore a 1e-3 non avrá senso spero
#path = '../model_zero/' #con regolarizzatore a zero
#path = '../model_alphae6/'  # con regolarizzatore a 1e-6

t = 1900
```

```python
loss_list = pd.read_csv(path + 'data_'+str(t)+'epochs'+'.csv', header=0, index_col=False, squeeze=False)
loss_list.pop('Unnamed: 0')
loss_list
```

```python
xsize = 8
ysize = 5

#fig, ax = plt.subplots(figsize=(xsize, ysize))

plt.figure(figsize=(xsize, ysize))
for key in loss_list.keys():
    plt.semilogy(loss_list[key],label = key)
    #plt.plot(loss_list[key],label = key)

plt.xlabel('epochs', labelpad = 10)
plt.savefig(path + 'plot/loss'+'.png', dpi = 300, pad_inches = .1, bbox_inches = 'tight')
plt.legend()
```
